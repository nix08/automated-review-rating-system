```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
# Sample DataFrame
np.random.seed(42)
data = pd.DataFrame({
    'Review Rating': np.random.choice([1, 2, 3, 4, 5], size=100),
    'Sentiment Score': np.random.normal(loc=0.5, scale=0.2, size=100)
})

# Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Review Rating', y='Sentiment Score', data=data, palette='coolwarm')
plt.title('Sentiment Score by Review Rating')
plt.xlabel('Review Rating')
plt.ylabel('Sentiment Score')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a built-in dataset
iris_df = sns.load_dataset("iris")

# Display the first few rows of the DataFrame
print("First 5 rows of the Iris dataset:")
print(iris_df.head())

# Create a simple scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris_df)
plt.title("Sepal Length vs. Sepal Width in Iris Dataset")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
df = sns.load_dataset("titanic")
```

```python
sns.histplot(df['age'].dropna())
plt.title("Histogram of Age")
plt.show()
```

```python
sns.countplot(x='class', data=df)
plt.title("Passenger Class Count")
plt.show()
```

```python
sns.barplot(x='sex', y='survived', data=df)
plt.title("Survival Rate by Sex")
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=df)
plt.title("Age Distribution by Class")
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=df)
plt.title("Violin Plot - Age by Class")
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=df)
plt.title("Age vs Fare")
plt.show()
```

```python
sns.swarmplot(x='class', y='age', data=df)
plt.title("Swarm Plot - Age by Class")
plt.show()
```

```python
sns.stripplot(x='class', y='age', data=df, jitter=True)
plt.title("Strip Plot - Age by Class")
plt.show()
```
