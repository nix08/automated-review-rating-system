```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python

x = np.linspace(0, 10, 100)
y = np.sin(x)


plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sine Wave', color='purple')
plt.title('Sample Sine Wave Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python

categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

```python
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

```python
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

```python
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
fig, axs = plt.subplots(2, 1)
axs[0].plot(x, np.sin(x))
axs[1].plot(x, np.cos(x))
plt.suptitle("Subplots Example")
plt.show()
```

```python
#Changing color and style
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title("Styled Line")
plt.show()
```

```python
#Adding labels and title
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Labeled Plot")
plt.show()
```

```python
#Fill between
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.title("Fill Between")
plt.show()
```

```python
#plt.plot(x, y)
plt.annotate("Peak", xy=(np.pi/2, 1), xytext=(2, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title("Annotation Example")
plt.show()
```
