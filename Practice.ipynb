```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

driver = webdriver.Chrome(executable_path='path/to/chromedriver')
driver.get("https://www.amazon.in/product-reviews/B07X1KT6BY")

reviews = driver.find_elements(By.CSS_SELECTOR, "[data-hook='review']")
data = []

for r in reviews:
    title = r.find_element(By.CSS_SELECTOR, "[data-hook='review-title']").text
    rating = r.find_element(By.CSS_SELECTOR, "[data-hook='review-star-rating']").text
    body = r.find_element(By.CSS_SELECTOR, "[data-hook='review-body']").text
    data.append({'Title': title, 'Rating': rating, 'Review': body})

df = pd.DataFrame(data)
df.to_csv('amazon_reviews.csv', index=False)
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Set headers to mimic Chrome browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/115 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9'
}

# Step 2: Define the Amazon review page URL
url = "https://www.amazon.in/ZEBRONICS-Zeb-Blitz-Headphone-Software-Simulated/dp/B09P1KK9K1/?_encoding=UTF8&pd_rd_w=2NhnC&content-id=amzn1.sym.3eb27980-84ee-4cfe-bafb-8634b1863e1c%3Aamzn1.symc.36bd837a-d66d-47d1-8457-ffe9a9f3ddab&pf_rd_p=3eb27980-84ee-4cfe-bafb-8634b1863e1c&pf_rd_r=MY7E30BK7F0QABSHDEQW&pd_rd_wg=mYIvN&pd_rd_r=e1518b04-7ec7-401b-beec-f705c04dd32d&ref_=pd_hp_d_btf_ci_mcx_mr_hp_atf_m&th=1"

# Step 3: Send request and parse HTML
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# Step 4: Extract review blocks
review_blocks = soup.find_all('div', {'data-hook': 'review'})

# Step 5: Extract review details
reviews = []
for block in review_blocks:
    title = block.find('a', {'data-hook': 'review-title'})
    rating = block.find('i', {'data-hook': 'review-star-rating'})
    body = block.find('span', {'data-hook': 'review-body'})
    
    reviews.append({
        'Title': title.text.strip() if title else '',
        'Rating': rating.text.strip() if rating else '',
        'Review': body.text.strip() if body else ''
    })

# Step 6: Save to CSV
df = pd.DataFrame(reviews)
df.to_csv('amazon_reviews.csv', index=False)
df.head()
```
